```python
# Chargement des donnees kc1
import pandas as pd
df1 = pd.read_csv("../data/kc1.csv")
df1.head()
```

```python
# Chargement des donnees kc2
df2 = pd.read_csv("../data/kc2.csv")
df2.head()
```

```python
df1.columns
```

```python
df2.columns
```

```python
print(df1.dtypes)
```

```python
print(df2.dtypes)
```

```python
# renommer la colonne "problems" en "defects"
df2 = df2.rename(columns={'problems': 'defects','lOCodeAndComment':'locCodeAndComment'})
df2.head()
```

```python
# transformer la colonne defects en booléen
df2['defects'] = df2['defects'].map({'yes': True, 'no': False})
```

```python
df2.head()
```

```python
# assembler les deux datasets
df = pd.concat([df1, df2], ignore_index=True)
```

```python
df.head()
```

```python
df.shape
```

```python
print(df.isna().sum())
```

```python
# On remplace les NaN par False, puis on convertit tout en 0 et 1
df['defects'] = df['defects'].fillna(False).astype(int)
```

```python
# Sauvegarde des donnees non equilibrees
import os
save_path = os.path.join('../data', 'cleaned_data.csv')
df.to_csv(save_path, index=False)
print(f"Fichier sauvegarde ici : {save_path}")
```

```python
# Analyse de la distribution avant equilibrage
import matplotlib.pyplot as plt

class_counts = df['defects'].value_counts().sort_index()
total = len(df)

print(f"Classe 0 (Sans Bug): {class_counts[0]} ({class_counts[0]/total*100:.2f}%)")
print(f"Classe 1 (Avec Bug): {class_counts[1]} ({class_counts[1]/total*100:.2f}%)")
print(f"Ratio: {class_counts[0] / class_counts[1]:.2f}:1")

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.bar(['Sans Bug', 'Avec Bug'], class_counts.values, color=['green', 'red'])
plt.ylabel('Nombre')
plt.title('Distribution des Classes')

plt.subplot(1, 2, 2)
plt.pie(class_counts.values, labels=['Sans Bug', 'Avec Bug'], autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Proportion')

plt.tight_layout()
plt.show()
```

```python
# Equilibrage des donnees par SMOTE
from imblearn.over_sampling import SMOTE

X = df.drop('defects', axis=1)
y = df['defects']

smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X, y)

df_balanced = pd.DataFrame(X_balanced, columns=X.columns)
df_balanced['defects'] = y_balanced

print(f"Avant equilibrage: {len(df)} lignes")
print(f"Apres equilibrage: {len(df_balanced)} lignes")
print(f"\nDistribution finale:")
print(df_balanced['defects'].value_counts())
```

```python
# Sauvegarde des donnees equilibrees
import os
save_path = os.path.join('../data', 'cleaned_data.csv')
df_balanced.to_csv(save_path, index=False)
print(f"Fichier equilibre sauvegarde dans {save_path}")
```

```python
# Visualisation apres equilibrage
class_counts_balanced = df_balanced['defects'].value_counts().sort_index()
total_balanced = len(df_balanced)

print(f"Dataset equilibre: {total_balanced} lignes")
print(f"Classe 0 (Sans Bug): {class_counts_balanced[0]} ({class_counts_balanced[0]/total_balanced*100:.1f}%)")
print(f"Classe 1 (Avec Bug): {class_counts_balanced[1]} ({class_counts_balanced[1]/total_balanced*100:.1f}%)")

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.bar(['Sans Bug', 'Avec Bug'], class_counts_balanced.values, color=['green', 'red'])
plt.ylabel('Nombre')
plt.title('Distribution des Classes APRES Equilibrage')

plt.subplot(1, 2, 2)
plt.pie(class_counts_balanced.values, labels=['Sans Bug', 'Avec Bug'], autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Proportion APRES')

plt.tight_layout()
plt.show()
```
